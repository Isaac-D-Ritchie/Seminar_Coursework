# Interactive Lab Task: Debug the Broken To-Do App

**Scenario:** A previous employee started building a To-Do List GUI but left several bugs and unfinished features. Your job is to run the code, investigate the problems, and fix them.

### Part I - Run the broken code

````python
import tkinter as tk

root = tk.Tk()
root.title("To Do Manager")
root.geometry("450x450")  # DOES THIS NEED ADJUSTMENT? - Changed from 200x200 to 450x450


tasks = []

def add_task():
    task = entry.get()

    if task == "":
        status_label.config(text=f"Task cannot be empty, current tasks: {len(tasks)}")
        return
    
    if task in tasks:
        status_label.config(text=f"Task already exists, current tasks: {len(tasks)}")
        return

    tasks.append(task)

    listbox.insert(0,task)   # BUG HERE - Added index

    entry.delete(0, tk.END)

    status_label.config(text=f"Task added, current tasks: {len(tasks)}")


def delete_task():

    selected = listbox.curselection()

    if not selected: #Added for error handling
        status_label.config(text=f"No task selected, current tasks: {len(tasks)}")

    task = listbox.get(selected)

    listbox.delete(selected)

    tasks.remove(task)

    status_label.config(text=f"Task removed, current tasks: {len(tasks)}")


def mark_done():

    selected = listbox.curselection()

    if not selected: #Added for error handling
        status_label.config(text=f"No task selected, current tasks: {len(tasks)}")

    task = listbox.get(selected)

    listbox.delete(selected)

    tasks.remove(task)

    listbox.insert(0,"[DONE]" + task)   # BUG HERE - Added index

    status_label.config(text=f"Task completed, current tasks: {len(tasks)}")


title = tk.Label(root, text="My To Do List", font=("Arial", 16))
title.pack(pady=10)

entry = tk.Entry(root, width=30)
entry.pack(pady=5)
entry.bind("<Return>", lambda e: add_task()) # Added for enter button detection

add_btn = tk.Button(root, text="Add Task", command=add_task)
add_btn.pack()

listbox = tk.Listbox(root, height=10)
listbox.pack(pady=10)

done_btn = tk.Button(root, text="Mark Done", command=mark_done)
done_btn.pack()

delete_btn = tk.Button(root, text="Delete Task", command=delete_task)
delete_btn.pack()

status_label = tk.Label(root, text="Ready")
status_label.pack(pady=10)

root.mainloop()
````
